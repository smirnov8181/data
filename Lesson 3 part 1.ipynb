```python
export_2016 = pd.read_excel('russia_2016.xlsx', sheet_name='Partner')
export_2017 = pd.read_excel('russia_2017.xlsx', sheet_name='Partner')
```

```python
export = pd.concat([export_2016, export_2017])
```

```python
export.sample(10)
```

```python
exclude = ['Europe & Central Asia', ' World', 'East Asia & Pacific', 'Middle East & North Africa', 'United States Minor Outlying I']
export_filter = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filter.pivot_table(index=['Year', 'Partner Name'],  values = 'Export (US$ Thousand)', aggfunc='sum')\
        .sort_values(['Year', 'Export (US$ Thousand)'], ascending=[True, False])
```

```python
export_pivot[export_pivot.index.get_level_values('Year') == 2016]
```
